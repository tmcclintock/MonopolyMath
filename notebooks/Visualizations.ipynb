# Visualizations

This notebook is for creating helpful figures to illustrate probabilities in Monopoly. For instance, visualizing the roll matrix, and transition probabilities.

```python
import numpy as np
import monopolymath as mm
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
#Figure formatting
#plt.rc("text", usetex=True)
plt.rc("font", size=12, family="serif")
```

```python
#Make a board object
mb = mm.MonopolyBoard()
mb.assign_dice() #Basic 2d6 dice
```

```python
R = mb.roll_matrix
print(R[0])
```

```python
x = np.arange(mb.number_of_spaces)
names = mb.space_names
fig, ax = plt.subplots(1,1)
im = ax.imshow(R)
ax.set_xticks(x)
_ = ax.set_xticklabels(names, fontsize=8, rotation='vertical')
ax.set_yticks(x)
_ = ax.set_yticklabels(names, fontsize=8)
cbar = fig.colorbar(im, ax=ax)
cbar.set_label("Transition Probability (2d6)")
```

```python
cmap = plt.cm.jet
cmaplist = [cmap(i) for i in range(cmap.N)]
cmap = cmap.from_list('Custom cmap', np.array(cmaplist)/36, cmap.N)
bounds = np.linspace(0,8./36.,9)
norm = mpl.colors.BoundaryNorm(bounds, cmap.N)
ax2 = fig.add_axes([0.95, 0.1, 0.03, 0.8])
cbar = mpl.colorbar.ColorbarBase(ax2, cmap=cmap, norm=norm, spacing='proportional', 
                                 ticks=bounds, boundaries=bounds, format='%.2f')
```
